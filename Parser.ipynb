```python
# functions for penalties extraction
def extract_penalties(sentence, name, i, decision_dic, nature, find_lst, addition_nature_lst=None):
    """ Extract one sentence of court decisions from decision column"""
    k = i + 1
    p = []
    while k < len(sentence):
        if addition_nature_lst is None:
            if (str(sentence.get(k).nature) == nature):
                break
            else:
                p.append(sentence.get(k).word)
        else:
            if (str(sentence.get(k).nature) == nature) or (str(sentence.get(k).nature) in addition_nature_lst):
                break
            else:
                p.append(sentence.get(k).word)
        k += 1
    if len(p) > 0:
        decision_dic[name] = "".join(p)
    find_lst[name] = False
    return find_lst, decision_dic

def clean_penalties(s, index):
    """find all penalties from a single sentence"""
    def find_penalties(decision):
        decision_dic = {}
        #and (("维持" not in decision)
        if (("被告" in decision) or ("上诉" in decision)) and ("撤销" not in decision):
            sentence = HanLP.segment(decision)
            find_lst = {"被告人":True, "上诉人":True, "被告":True, "被告单位":True, "有期徒刑":False, 
                        "罚金":False, "缓刑":False, "羁押":False, "拘役":False}
            for i in range(len(sentence) - 1):
                # ignore special words (维持，驳回，撤销)
                if (find_lst["被告单位"]) and (sentence.get(i+1).word == "单位"):
                    find_lst, decision_dic = extract_penalties(sentence, "被告单位", i+1, decision_dic, "v", find_lst)
                    find_lst.update({"被告人":False, "上诉人":False, "被告":False,
                                     "有期徒刑":True, "罚金":True, "缓刑":True, 
                                     "羁押":True, "拘役":True})
                if (find_lst["上诉人"]) and (sentence.get(i).word == "上诉人"):
                    find_lst, decision_dic = extract_penalties(sentence, "上诉人", i, decision_dic, "v", find_lst)
                    find_lst.update({"被告人":False, "被告单位":False, "被告":False,
                                     "有期徒刑":True, "罚金":True, "缓刑":True, 
                                     "羁押":True, "拘役":True})
                if (find_lst["被告人"]) and (sentence.get(i).word == "被告人"):
                    find_lst, decision_dic = extract_penalties(sentence, "被告人", i, decision_dic, "v", find_lst, ["vi", "vn", "n"])
                    find_lst.update({"被告单位":False, "上诉人":False, "被告":False,
                                     "有期徒刑":True, "罚金":True, "缓刑":True, 
                                     "羁押":True, "拘役":True})
                if (find_lst["被告"]) and (sentence.get(i).word == "被告"):
                    find_lst, decision_dic = extract_penalties(sentence, "被告", i, decision_dic, "v", find_lst)
                    find_lst.update({"被告人":False, "上诉人":False, "被告单位":False,
                                     "有期徒刑":True, "罚金":True, "缓刑":True, 
                                     "羁押":True, "拘役":True})
                if (find_lst["有期徒刑"]) and (sentence.get(i).word == "有期徒刑"):
                    find_lst, decision_dic = extract_penalties(sentence, "有期徒刑", i, decision_dic, "w", find_lst)
                if (find_lst["罚金"]) and (sentence.get(i).word == "罚金"):
                    find_lst, decision_dic = extract_penalties(sentence, "罚金", i, decision_dic, "w", find_lst)
                if (find_lst["罚金"]) and (sentence.get(i).word == "处罚金"):
                    find_lst, decision_dic = extract_penalties(sentence, "罚金", i, decision_dic, "w", find_lst)
                if (find_lst["缓刑"]) and (sentence.get(i).word == "缓刑"):
                    find_lst, decision_dic = extract_penalties(sentence, "缓刑", i, decision_dic, "w", find_lst, ["vn"])    
                if (find_lst["羁押"]) and (sentence.get(i).word == "羁押"):
                    find_lst, decision_dic = extract_penalties(sentence, "羁押", i, decision_dic, "w", find_lst, ["ude1"])    
                if (find_lst["拘役"]) and (sentence.get(i).word == "拘役"):
                    find_lst, decision_dic = extract_penalties(sentence, "拘役", i, decision_dic, "w", find_lst, ["v"])    
        return decision_dic
    
    # reformat the decision list
    index_to_clean = index_clean()
    if index not in index_to_clean:  
        lst = ast.literal_eval(s)
        sentence_lst = []
        splitted = []
        for i in lst:
            segment = i.split("。")
            for y in segment:
                splitted.append(y)
        for i in splitted:
            seg = i.split("；")
            for j in seg:
                sentence_lst.append(j)
    else:
        sentence_lst = []
        splitted = []
        splitted = s.split("。")
        for i in splitted:
            o = i.split("；")
            for j in o:
                sentence_lst.append(j)
    penalties = []
#     if ("维持" in sentence) or ("撤销" in sentence) or ("免予" in test):
#         return {}
#     if ("维持" in s) or ("撤销" in s):
#          return {}
    for i in sentence_lst:
        p = find_penalties(i)
        if "免予" in i:
            p["免予处罚"] = True
        if len(p) > 1:
            penalties.append(p)
    return penalties
# penalties = clean_penalties(sentence1, -1)
# penalties
```
